# R

This dataset represents student grades across different courses (ADM, LAW, ECO), showing their scores in three assessments (AP1, AP2, AP3), their calculated Grade, and the corresponding letter Score.

```r
grades<-read.csv("https://raw.githubusercontent.com/lcbjrrr/data/main/grades%20-%20okk.csv")
head(grades,3)
```

Let's create a histogram of the "Grade" column from the grades data frame, using specified breaks (0, 5, 7, 9, 10) to define the bins, and displays the frequency (counts) of grades within each bin.

```r
hist(grades$Grade,breaks=c(0,5,7,9,10),freq=T )
```

Let's calculate and display the summary statistics (minimum, first quartile, median, mean, third quartile, and maximum) of the "Grade" column from the grades data frame.

```r
summary(grades$Grade)
```

Let's create a boxplot to visually represent the distribution and potential outliers of the "Grade" column.

```r
boxplot(grades$Grade)
```

Let's create a scatter plot, visualizing the relationship between the "AP2" and "AP3" columns from the grades data frame. Each point on the plot represents a student's score in AP2 and AP3.

```r
plot(grades$AP2,grades$AP3)
```
