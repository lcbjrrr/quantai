Let's defines a new function. One to check the approval of student that returns "PASS" if nota is 7 or greater and freq is 0.75 or greater, otherwise it returns "FAIL", and then it applies this function to the values 9 and 0.8, printing the result ("PASS").

```r
approval <-function(nota, freq){
    if (nota>=7 & freq>=0.75){
        aprov <- "PASS"
    }else{
        aprov <- "FAIL"
    }
    return(aprov)
}
john<-approval(9,0.8)
print(john)
```

Let's define a function avg() that calculates and returns the average of a fixed vector of grades, and then prints the result.

```r
avg <- function(){
    grades <- c(5,5,6,6,7,7,8,8,9,9)
    s<-0
    for (n in grades ){
      s<-s+n
    }
    a<- s /10
    return(a)
}

aa <- avg()
print(aa)
```

Let's iterate through the vector's indices, and then prints the result.

```r
avg <- function(){
    grades <- c(5,5,6,6,7,7,8,8,9,9)
    s<-0
    for (i in 1:10 ){
      s <- s + grades[i]
    }
    a<- s /10
    return(a)
}

aa <- avg()
print(aa)
```

This dataset shows the daily count of a specific event (likely new cases of a disease) for Brazil, India, and the US, starting from January 23, 2020.

```r
covid <- read.csv('https://raw.githubusercontent.com/lcbjrrr/quant/master/covid19%20-%20cases.csv')
head(covid,3)
```

Let's convert the "date" column in the covid dataframe to date objects and then uses those dates as row names.

```r
covid$date <-  as.Date(covid$date, "%m/%d/%y")
rownames(covid) <- covid$date
head(covid,3)
```

Let's create a line plot showing the trend of "Brazil" values against "date" from the dataframe "covid"

```r
plot(Brazil~date,data=covid,type='l')
```

Let's create now a line plot showing the trend of the "Brazil" column over "date" in red, and then adds lines for "India" (blue) and "US" (green) to the same plot, visualizing the comparative trends of these three countries over time.

```r
plot(covid$date, covid$Brazil, type = "l", col = "red")
lines(covid$date, covid$India, type = "l", col = "blue")
lines(covid$date, covid$US, type = "l", col = "green")
```
