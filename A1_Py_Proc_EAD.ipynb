# Data Manipulation


A CSV (Comma-Separated Values) file is a simple text format used to store tabular data, where each row represents a record and columns are separated by commas. It is commonly used for data exchange between applications. A DataFrame is a data structure in pandas, a Python library, that allows you to store and manipulate structured data in a table-like format. In the code auto = pd.read_csv("https://raw.githubusercontent.com/lcbjrrr/quantai/refs/heads/main/autoinsurance.csv"), pd.read_csv() loads the CSV file from a URL into a DataFrame called auto. You can then perform various data operations, such as analysis or transformation, on the DataFrame.

```python
import pandas as pd
auto = pd.read_csv("https://raw.githubusercontent.com/lcbjrrr/quantai/refs/heads/main/autoinsurance.csv")
auto
```

The dataset contains auto insurance data for different U.S. states, including columns for state abbreviation, capital, price level, insurance premiums for the years 2021–2023, population, geographic coordinates (latitude and longitude), and regional information. The data provides insights into how premiums vary by state, population size, and region, allowing for analysis of trends and patterns in auto insurance across the country.


The function auto.head(3) displays the first three rows of the DataFrame auto, showing the top entries in the dataset for a quick preview of the data.

```python
auto.head(3)
```

auto.tail(2) returns the last two rows of the auto DataFrame, showing the most recent records from the dataset.

```python
auto.tail(2)
```

To select a column in a pandas DataFrame, you can use the column name in square brackets, like auto['Y2023']. This retrieves the data for the "Y2023" column, showing the insurance premiums for that year across all rows in the dataset.

```python
auto['Y2023']
```

To select multiple columns from a pandas DataFrame, you can pass a list of column names inside double square brackets, like this: `auto[['Y2023', 'Y2022', 'Y2021']]`. This will return a new DataFrame containing only the specified columns "Y2023," "Y2022," and "Y2021."

```python
auto[['Y2023','Y2022','Y2021']]
```


To operate with columns in a pandas DataFrame, you can perform arithmetic operations directly between columns. For example, avg = (auto['Y2023'] + auto['Y2022'] + auto['Y2021']) / 3 calculates the average of the "Y2023", "Y2022", and "Y2021" columns for each row in the DataFrame. The result is a new Series containing the average premium for each state or record. You can assign this to a new column in the DataFrame if needed: auto['Average'] = avg.

```python
avg = (auto['Y2023'] + auto['Y2022'] + auto['Y2021'])/3
avg
```

To filter rows in a pandas DataFrame based on a specific condition, like selecting rows where the "Region" column is equal to "Midwest", you can use:

```python
auto[auto['Region']=='Midwest']
```

To filter rows where the "Y2023" column has values greater than 999, use the following code:

```python
auto[auto['Y2023']>999]
```

# Summarization / Aggregation


Summarization like a pivot table is used to group data by a specific column (e.g., "Region") and calculate summary statistics (e.g., average) for each group. In avg_premium_region = auto.groupby('Region')['Y2023'].mean(), the data is grouped by the "Region" column, and the mean of the "Y2023" column (average premium for 2023) is calculated for each region. This creates a summary that shows the average premium for each region in the dataset.

```python
avg_premium_region = auto.groupby('Region')['Y2023'].mean()
avg_premium_region
```


A bar chart serves as a visual tool for comparing data across different categories. In avg_premium_region.plot(kind='bar'), it creates a bar chart where the x-axis represents the regions, and the y-axis shows the average premium for 2023 (Y2023). Each bar’s height represents the average premium for a region, making it easy to visually compare premium values across different regions and quickly identify trends or differences.

# Data Visualization

```python
avg_premium_region.plot(kind='bar')
```

To compare multiple columns, avg_premium_region = auto.groupby('Region')[['Y2023', 'Y2022', 'Y2021']].mean() calculates the average premium for each year ("Y2023", "Y2022", "Y2021") grouped by the "Region". This provides a summary of the average premiums across different regions for the three years, allowing you to compare trends and differences in premiums over time within each region.

```python
avg_premium_region = auto.groupby('Region')[['Y2023','Y2022','Y2021']].mean()
avg_premium_region
```

Pandas integrates with Matplotlib for data visualization. When you use avg_premium_region.plot(kind='bar'), pandas automatically calls Matplotlib to create the bar chart. This allows you to visualize the average premiums for different regions. The plot() function is a convenient wrapper that leverages Matplotlib's plotting capabilities without needing to write additional Matplotlib code. This results in an easy-to-use method for visualizing and comparing data in a DataFrame.

```python
avg_premium_region.plot(kind='bar')
```

In pandas, aggregation functions like count() can be used to count occurrences of non-null values in a column. In the example count_state_region = auto.groupby('Region')['state'].count(), the data is grouped by "Region," and the count() function counts the number of non-null values in the "state" column for each region. This helps you determine how many states are represented in each region of the dataset.

```python
count_state_region = auto.groupby('Region')['state'].count()
count_state_region
```

A pie chart is used to visualize the proportion of categories within a dataset. When you run count_state_region.plot(kind='pie'), it creates a pie chart based on the counts of states in each region. Each slice of the pie represents a region, and the size of each slice corresponds to the number of states in that region. This provides an intuitive way to compare the relative frequencies of states across different regions.

```python
count_state_region.plot(kind='pie')
```
