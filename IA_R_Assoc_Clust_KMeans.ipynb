The cluster package in R provides tools for cluster analysis, including various clustering algorithms and functions for evaluating clustering results.

```r
install.packages("cluster")
library(cluster)
```

This dataset shows two numerical variables: "salario" (salary) and "saldo" (balance), likely representing financial data for individuals.

```r
salario_conta <- read.csv('https://raw.githubusercontent.com/lcbjrrr/data/main/salario_conta%20.csv')
print(salario_conta)
```

Let's create a scatter plot where "salario" (salary) is plotted against "saldo" (balance) from the salario_conta data frame, visualizing the relationship between these two financial variables.

```r
plot(salario_conta$salario ~ salario_conta$saldo)
```

Let's perform a k-means clustering on the salario_conta dataset, dividing it into 5 clusters, displays the cluster assignments, and then visualizes the clustering results.

```r
k <- kmeans(x = salario_conta,centers = 5)
print(k$cluster)
clusplot(x = salario_conta, clus = k$cluster, color = T, shade = T, labels = 5, lines = 0,main = "KMEANS")
```

Let's perform k-means clustering on the volei dataset, aiming to group the data into 2 clusters. It then prints the cluster assignments for each data point and visualizes the resulting clusters using a cluster plot.

```r
volei <- read.csv('https://raw.githubusercontent.com/lcbjrrr/data/main/volei%20-%20ok.csv')
head(volei,1)
```

```r
k <- kmeans(x = volei,centers = 2)
print(k$cluster)
clusplot(x = volei, clus = k$cluster, color = T, shade = T, labels = 2, lines = 1,main = "KMEANS")
```
